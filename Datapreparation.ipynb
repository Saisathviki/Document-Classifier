## Data Preprocessing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# Load AG News dataset
train_df = pd.read_csv('ag_news/train.csv', header=None, names=['class', 'title', 'description'])
test_df = pd.read_csv('ag_news/test.csv', header=None, names=['class', 'title', 'description'])

# Combine title and description
train_df['text'] = train_df['title'] + ' ' + train_df['description']
test_df['text'] = test_df['title'] + ' ' + test_df['description']

# Class mapping
class_map = {
    1: "World",
    2: "Sports",
    3: "Business",
    4: "Science/Tech"
}

# Visualize class distribution
plt.figure(figsize=(10, 5))
sns.countplot(x='class', data=train_df)
plt.title('Class Distribution')
plt.xticks(ticks=[0,1,2,3], labels=class_map.values())
plt.show()
```
